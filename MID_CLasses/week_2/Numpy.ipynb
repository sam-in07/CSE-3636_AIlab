```python
# NumPy example code:

import numpy as np

# Step 1: Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Step 2: Perform operations
print("Original array:", arr)
print("Add 5 to each element:", arr + 5)
print("Square each element:", arr ** 2)
print("Sum of all elements:", np.sum(arr))
```
