A set is a collection which is unordered, unchangeable*, and unindexed.Sets are written with curly brackets.Set items are unordered, unchangeable, and do not allow duplicate values.Once a set is created, you cannot change its items, but you can remove items and add new items.



```python
thisset = {"apple", "banana", "cherry", "apple"}

print(thisset)
```

```python
set1 = {"abc", 34, True, 40, "male"}
print(set1)
```

To determine how many items a set has, use the len() function.


It is also possible to use the set() constructor to make a set.

```python
thisset = set(("apple", "banana", "cherry"))
print(thisset)
```

Set doesn't have access by number index.Set do not allow duplicate values. When we include duplicate value to a set, then the duplicate is ignored and set remains unchanged.

To add one item to a set use the add() method

```python
thisset = {"apple", "banana", "cherry"}

thisset.add("orange")

print(thisset)
```
