Scikit-learn Example Code

```python
from sklearn.linear_model import LinearRegression

# Step 1: Define the data
# X = input numbers, y = output numbers
X = [[1], [2], [3], [4], [5]]
y = [2, 4, 6, 8, 10]  # Pattern: y = 2 * x

# Step 2: Create the model
model = LinearRegression()

# Step 3: Train the model
model.fit(X, y)

# Step 4: Make predictions
prediction = model.predict([[6]])  # What comes after 5?
print("Prediction for input 6:", prediction)
```

Matplotlib Example Code


```python
import matplotlib.pyplot as plt

# Step 1: Define data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]  # Pattern: y = 2 * x

# Step 2: Create the plot
plt.plot(x, y)

# Step 3: Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Step 4: Show the plot
plt.show()

```


NumPy Example Code







```python

import numpy as np

# Step 1: Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Step 2: Perform operations
print("Original array:", arr)
print("Add 5 to each element:", arr + 5)
print("Square each element:", arr ** 2)
print("Sum of all elements:", np.sum(arr))
```

Pandas Example Code

```python

import pandas as pd

# Step 1: Create a DataFrame
data = pd.DataFrame({
	'Name': ['Alice', 'Bob', 'Charlie'],
	'Age': [25, 30, 22],
	'City': ['New York', 'London', 'Paris']
})

# Step 2: Display the DataFrame
print("DataFrame:")
print(data)

# Step 3: Access a column
print("\nAges:")
print(data['Age'])

# Step 4: Filter rows
print("\nPeople older than 23:")
print(data[data['Age'] > 23])
```
