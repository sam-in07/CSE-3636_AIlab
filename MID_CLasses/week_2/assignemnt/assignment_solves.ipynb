```

scores = (85, 90, 78, 92, 88, 76)


highest = max(scores)
lowest = min(scores)


average = sum(scores) / len(scores)


print("Highest Score:", highest)
print("Lowest Score:", lowest)
print("Average Score:", round(average, 2))
```

```
students = {"Alice": 85, "Bob": 78, "Charlie": 92}


top_student = max(students, key=students.get)
top_marks = students[top_student]

print("Top Student:", top_student, "with", top_marks, "marks")


name = input("Enter a student name: ")

if name in students:
    print(name, "scored", students[name])
else:
    print("Student not found")
```

```
fruits = {"Alice": "Apple", "Bob": "Banana", "Charlie": "Grapes"}


for student, fruit in fruits.items():
    print(student, "likes", fruit)


fruits["David"] = "Mango"
print("\nNew student added: David likes Mango")


name = input("Enter a student name: ")

if name in fruits:
    print(name, "likes", fruits[name])
else:
    print("Student not found")
```
