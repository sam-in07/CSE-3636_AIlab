```
def guess_password():
    secret_password = "1234"  # Secret password
    attempts = 0

    # Loop through all possible 4-digit combinations
    for i in range(10000):
        attempt = str(i).zfill(4)  # Format the number to be a 4-digit string
        attempts += 1
        if attempt == secret_password:
            print(f"Password found: {attempt}")
            print(f"Number of attempts: {attempts}")
            break

guess_password()
```

```
def is_prime(n) :
  if n<=1 :
    return False
  for i in range (2 , int(n**0.5)+1) :
    if(n%i) == 0 :
      return False

  return True


n = int(input("Enter a number : "))
if(is_prime(n)) :
  print(f"{n} is a prime number")
else :
  print(f"{n} is not a prime number")


```

Problem 3



```
def find_largest():
    numbers = []
    for _ in range(5):
        numbers.append(int(input("Enter a number: ")))

    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num

    print(f"The largest number is: {largest}")

find_largest()
```

# Question 4 – Find Pair with Target Sum

```
def find_pair_with_sum(numbers, target):
    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] + numbers[j] == target:
                print(f"Pair found: {numbers[i]} and {numbers[j]}")
                return
    print("No pair found.")

numbers = [2, 4, 7, 11, 15]
target = 9
find_pair_with_sum(numbers, target)
```

Question 5 – Count Number Frequency


```
def count_frequency(numbers):
    frequency = {}
    for num in numbers:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1
    print(frequency)

numbers = [3, 5, 3, 7, 9, 5, 3]
count_frequency(numbers)
```

### Question 7 – Grade Calculator


```
def grade_calculator(marks):
    if 80 <= marks <= 100:
        return 'A'
    elif 70 <= marks <= 79:
        return 'B'
    elif 60 <= marks <= 69:
        return 'C'
    elif 50 <= marks <= 59:
        return 'D'
    else:
        return 'F'

marks = int(input("Enter your marks: "))
grade = grade_calculator(marks)
print(f"Your grade is: {grade}")
```

Question 9 – Student DataFrame Analysis


```
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Marks': [85, 72, 90, 65, 78]
}

df = pd.DataFrame(data)
print(df)

average_marks = df['Marks'].mean()
print(f"Average Marks: {average_marks}")

above_75 = df[df['Marks'] > 75]
print("Students with marks above 75:")
print(above_75)
```

### Question 11 – Linear Regression Prediction

```
from sklearn.linear_model import LinearRegression

# Training data
x = [[1], [2], [3], [4], [5]]
y = [2, 4, 5, 4, 5]

# Train the model
model = LinearRegression()
model.fit(x, y)

# Predict for x = 6
predicted_value = model.predict([[6]])
print(f"Predicted value: {predicted_value[0]}")

# Intercept and slope
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")
```


### Question 12 – Regression Visualization

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate random data
x = np.random.rand(100, 1) * 10  # 100 random numbers between 0 and 10
y = 2 * x + np.random.randn(100, 1) * 2  # Linear relationship with noise

# Scatter plot
plt.scatter(x, y, color='blue')

# Train the linear regression model
model = LinearRegression()
model.fit(x, y)

# Plot the regression line
y_pred = model.predict(x)
plt.plot(x, y_pred, color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regression Line')
plt.show()
```
