**Some built-in functions in Python**

```python
# Python script with built-in function examples

# 1. print(): This function prints the specified message to the console.
print("Hello, World!")  # Output: Hello, World!

# 2. len(): This function returns the length (number of items) of an object, such as a list, string, or tuple.
my_string = "Python"
print(len(my_string))  # Output: 6, as "Python" has 6 characters.

# 3. type(): This function returns the type of an object (e.g., string, integer, list, etc.).
print(type(my_string))  # Output: <class 'str'> because "Python" is a string.

# 4. int(): Converts a string or float to an integer (if possible).
my_float = 3.5
print(int(my_float))  # Output: 3 (float truncated to integer)

# 5. str(): Converts a number or other data types into a string.
my_num = 10
print(str(my_num))  # Output: '10', the number is now a string.

# 6. sum(): Adds all the items in an iterable (e.g., a list or tuple) and returns the sum.
my_list = [1, 2, 3, 4, 5]
print(sum(my_list))  # Output: 15 (sum of all the elements in the list)

# 7. max(): Returns the largest item in an iterable or multiple arguments.
print(max(my_list))  # Output: 5 (largest number in the list)

# 8. min(): Returns the smallest item in an iterable or multiple arguments.
print(min(my_list))  # Output: 1 (smallest number in the list)

# 9. input(): Accepts input from the user as a string. The input() function always returns a string
user_input = input("Enter your name: ")  # This will prompt the user to input their name.
print("Hello, " + user_input + "!")  # Example output: "Hello, John!" (if user entered "John")

# 10. abs(): Returns the absolute value of a number (removes any negative sign).
negative_num = -10
print(abs(negative_num))  # Output: 10
```

**Python Classes and Inheritance(the examples here are taken from geeksforgeeks website)**

Creating a class

```python
class Person():

  # Constructor

#(The __init__ method is a special method in Python, also known as the constructor. It is automatically called when a new object of the class is created.)
#self is a reference to the current object being created. It allows you to access attributes and methods of the current object.)

  def __init__(self, name, id):
    self.name = name
    self.id = id

  # To check if this person is an employee
  def Display(self):
    print(self.name, self.id)


# Driver code
emp = Person("Satyam", 102) # An Object of Person
emp.Display()
```


Creating a child class

The self parameter is mandatory in all instance methods in a class in python, whether it's the parent class or the child class.

self refers to the current instance of the class, which allows you to access instance variables and methods from within the class.
If you don't use self, Python won't know which object you're referring to, and you'll get an error. But we do not need self in some special cases such as static methods etc.

```python
class Emp(Person):

  def Print(self):
    print("Emp class called")

new_emp = Emp("Mayank", 103)

# calling parent class function
new_emp.Display()

# Calling child class function
new_emp.Print()
```
